```python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12, 8)  # Adjusts the configuration of the plots we will create

# Read in data (replace 'your_file.csv' with the actual file name)
df = pd.read_csv('/Users/anthonylaite/Desktop/projects/jobs.csv')

# Drop the unnamed column
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

# Display the entire DataFrame
df

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define the order of experience levels
experience_order = ['Entry-level', 'Mid-level', 'Senior', 'Executive']

# Boxplot for experience_level vs salary_in_usd with custom order
plt.figure(figsize=(12, 8))
sns.boxplot(x='experience_level', y='salary_in_usd', data=df, order=experience_order)
plt.title('Boxplot: Experience Level vs. Salary')
plt.xlabel('Experience Level')
plt.ylabel('Salary in USD')
plt.show()






```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a pivot table
pivot_table = df.pivot_table(values='salary_in_usd', index='job_category', aggfunc='mean')

# Plot the heatmap
sns.heatmap(pivot_table, cmap='YlGnBu', annot=True, fmt=".2f", cbar_kws={'label': 'Mean Salary (USD)'})
plt.title('Heatmap: Mean Salary vs. Job Category')
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select relevant columns for correlation
correlation_cols = ['salary_in_usd', 'job_count', 'mean_salary', 'min_salary', 'max_salary']

# Create a correlation matrix
correlation_matrix = df[correlation_cols].corr()

# Plot the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming these columns are relevant to your analysis
columns_of_interest = [
 'job_category', 'salary_in_usd',
 'employment_type', 'work_setting', 'company_location'
]  # Add the closing square bracket here

# Select only the columns of interest
numeric_columns = df[columns_of_interest]

# Calculate the correlation matrix
correlation_matrix_pearson = numeric_columns.corr()

# Display the correlation matrix
print(correlation_matrix_pearson)

# Create a heatmap with a custom color map and title
plt.figure(figsize=(12, 10))  # Adjust the figure size if needed
sns.heatmap(correlation_matrix_pearson, annot=True, cmap="RdBu_r", fmt=".2f")
plt.title("Pearson Correlation Test", fontsize=16)  # Add the title

plt.show()
```
